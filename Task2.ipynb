# 5. Finding and removing the duplicates

```python
import pandas as pd
if 'EXT_SEC_ID' in cs_fund_cleaned.columns:
    print("\nAnalyzing 'EXT_SEC_ID' column for duplicates...")

    duplicate_mask = cs_fund_cleaned['EXT_SEC_ID'].duplicated(keep=False)
    duplicate_count = cs_fund_cleaned['EXT_SEC_ID'].duplicated().sum()
    print(f"Number of duplicate entries in 'EXT_SEC_ID': {duplicate_count}")

    if duplicate_count > 0:
        duplicate_values = cs_fund_cleaned.loc[duplicate_mask, 'EXT_SEC_ID'].drop_duplicates().reset_index(drop=True)
        print("\nDuplicate values in 'EXT_SEC_ID':")
        print(duplicate_values)
        output_file = r"C:\Users\Neha\Desktop\OP_A2.xlsx"
        with pd.ExcelWriter(output_file, engine='xlsxwriter') as writer:
            duplicate_values.to_excel(writer, sheet_name='Duplicates', index=False, header=['Duplicate EXT_SEC_ID'])

            workbook = writer.book
            worksheet_duplicates = writer.sheets['Duplicates']

            red_format = workbook.add_format({'font_color': 'red', 'bg_color': '#FFC7CE'})
            worksheet_duplicates.conditional_format(
                f'A2:A{len(duplicate_values) + 1}',  # Highlight the duplicate ACCT_CD column
                {'type': 'formula', 'criteria': 'TRUE', 'format': red_format}
            )

        print(f"\nExcel file saved with duplicate values at {output_file}")
    else:
        print("\nNo duplicates found in 'EXT_SEC_ID'.")
else:
    print("\n'EXT_SEC_ID' column not found in the DataFrame.")
```

# 6. Datatype Suggestion

```python
import pandas as pd
def suggest_datatype(series):
    if pd.api.types.is_integer_dtype(series):
        return 'INTEGER'
    elif pd.api.types.is_float_dtype(series):
        return 'FLOAT'
    elif pd.api.types.is_datetime64_any_dtype(series):
        return 'DATETIME'
    elif pd.api.types.is_string_dtype(series):
        return 'VARCHAR(MAX)'
    else:
        return 'VARCHAR(MAX)'
filtered_columns = cs_fund_cleaned.columns.tolist()

data_types = {col: suggest_datatype(cs_fund_cleaned[col]) for col in filtered_columns}

print(data_types)
```
